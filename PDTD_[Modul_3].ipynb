**1. If, else**

```python
#pernyataan kondisional sederhana (2 kondisi)
cuaca = input("Apakah sedang hujan? (ya/tidak): ").lower()
if cuaca == "ya":
    print("Pakai payung!")
else:
    print("Tidak perlu pakai payung!")
```

**2. if, elif, else**

```python
#pernyataan kondisional dengan 3 kondisi
angka = int(input("Masukkan sebuah angka: "))  #inputan
if angka > 0:
    print("Angka tersebut positif.")
elif angka < 0:
    print("Angka tersebut negatif.")
else:
    print("Angka tersebut nol.")
```

if statement: Menjalankan blok kode jika kondisi benar.

elif statement: Memeriksa kondisi tambahan jika kondisi sebelumnya salah.

else statement: Menjalankan blok kode jika semua kondisi sebelumnya salah.

```python
print("üìñ === Perpustakaan Santai === üìñ")

# Inisialisasi variabel
stok_buku = 5

# Input jumlah peminjam
jumlah_peminjaman = int(input("Masukkan jumlah buku yang ingin dipinjam: "))

# Cek apakah peminjaman buku dapat dilakukan
if jumlah_peminjaman <= stok_buku:
    print("üéâ Peminjaman berhasil! Buku siap menemanimu membaca üìö‚ú®")
    stok_buku -= jumlah_peminjaman
    print("Sisa stok buku sekarang:", stok_buku)
else:
    print("üòÖ Peminjaman gagal. Stok nggak cukup nih!")
    print("Stok tersedia cuma:", stok_buku)
```

```python
print("üéüÔ∏è === Tiket Seru Bersama Teman === üéâ")

# Inisialisasi variabel
harga_tiket = 150000
jumlah_penumpang = int(input("Masukkan jumlah penumpang: "))
total_harga = harga_tiket * jumlah_penumpang

# Diskon untuk jumlah penumpang lebih dari 4
if jumlah_penumpang > 4:
    diskon = 0.1  # 10%
    total_harga_diskon = total_harga * (1 - diskon)
    print(f"Yeay! Kamu dan teman-teman mendapatkan diskon 10%! üéä")
    print(f"Total Harga Setelah Diskon: Rp {total_harga_diskon}")
else:
    print("Hmm, belum cukup rame untuk diskon üòÖ")
    print(f"Total Harga: Rp {total_harga}")

print("Selamat menikmati perjalanan! ‚úàÔ∏èüçø")
```

```python
print("=== CEK TINGKAT STRES SEDERHANA ===")

# Pertanyaan 1
p1 = input("1. Apakah kamu sering merasa cemas tanpa alasan jelas? (ya/tidak): ")

# Pertanyaan 2
p2 = input("2. Apakah kamu sulit tidur akhir-akhir ini? (ya/tidak): ")

# Pertanyaan 3
p3 = input("3. Apakah kamu merasa mudah marah atau tersinggung? (ya/tidak): ")

# Pertanyaan 4
p4 = input("4. Apakah kamu merasa lelah secara emosional? (ya/tidak): ")

# Hitung jumlah 'ya'
skor = 0
if p1.lower() == "ya":
    skor += 1

if p2.lower() == "ya":
    skor += 1

if p3.lower() == "ya":
    skor += 1

if p4.lower() == "ya":
    skor += 1

print("\n=== HASIL ===")

# Penilaian berdasarkan skor
if skor == 0:
    print("Kamu tampak sangat tenang. Terus jaga keseimbangan hidupmu!")
elif skor == 1:
    print("Ada sedikit tanda stres, tapi masih ringan. Tetap perhatikan keseharianmu ya.")
elif skor == 2:
    print("Kamu mungkin mengalami stres tingkat sedang. Cobalah istirahat lebih banyak.")
elif skor == 3:
    print("Tanda stresmu cukup jelas. Cobalah berbicara dengan seseorang atau lakukan kegiatan relaksasi.")
else:  # skor == 4
    print("Kamu tampak sangat stres. Jangan ragu mencari dukungan atau bercerita pada orang terpercaya.")
```

```python
print("=== Kalkulator Keuangan ===")

# Input data
pendapatan_harian = float(input("Masukkan pendapatan per hari (Rp): "))
hari_kerja = int(input("Masukkan jumlah hari kerja dalam seminggu: "))
pengeluaran_harian = float(input("Masukkan pengeluaran rata-rata per hari (Rp): "))

# Proses perhitungan
total_pendapatan = pendapatan_harian * hari_kerja
total_pengeluaran = pengeluaran_harian * hari_kerja
tabungan = total_pendapatan - total_pengeluaran

# Output utama
print("\nüíµ Total pendapatan mingguan:", total_pendapatan, "rupiah")
print("üí∏ Total pengeluaran mingguan:", total_pengeluaran, "rupiah")
print("üè¶ Total tabungan kamu minggu ini:", tabungan, "rupiah")

# Analisis sederhana dengan nested if-else
if tabungan > 1000000:
    print("\nüí∞ Hebat! Kamu mengelola uangmu dengan sangat baik.")
else:
    if tabungan > 0:
        print("\nüìä Kamu masih bisa menabung sedikit, tapi coba kurangi pengeluaran agar lebih efisien.")
    else:
        print("\n‚ö†Ô∏è Pengeluaran kamu melebihi pendapatan! Saatnya evaluasi gaya hidup dan anggaran.")
```

**Rumus** **Kabisat**

Jika tahun itu dibagi 4, maka itu adalah tahun kabisat, tetapi ada pengecualian...

Jika tahun itu juga dibagi 100, maka tahun itu TIDAK kabisat, kecuali...

Jika tahun itu dibagi 400, maka tahun itu tetap kabisat.
Jadi, untuk memudahkan:

Tahun yang bisa dibagi 4 adalah kabisat, kecuali kalau itu juga bisa dibagi 100, tetapi jika bisa dibagi 400, tetap kabisat.
Contoh:

Tahun 2020: Bisa dibagi 4, jadi kabisat.

Tahun 1900: Bisa dibagi 100, tapi tidak bisa dibagi 400, jadi bukan kabisat.

Tahun 2000: Bisa dibagi 400, jadi tahun kabisat.
